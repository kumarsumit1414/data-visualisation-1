```python
days=list(range(0,22,3))
temp=list(np.random.randint(30,50,8))
```

```python
plt.plot(days,temp)
```

```python
plt.scatter(days,temp,)
```

```python
temp_min=list(np.random.randint(30,50,8))
temp_max=list(np.random.randint(30,50,8))
plt.plot(days,temp_max,'or')
plt.plot(days,temp_min,'ob')
plt.plot(days,temp_max,'r')
plt.plot(days,temp_min,'b')
plt.xlabel('days')
plt.ylabel('temperature')
ax=plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xticks(range(0,28,4),['w1','w2','w3','w4','w5','w6','w7'])

plt.show()
```

```python
temp_min=list(np.random.randint(30,50,8))
temp_max=list(np.random.randint(30,50,8))
plt.axes().set_aspect('equal')
plt.plot(days,temp_max,'or')
plt.plot(days,temp_min,'ob')
plt.plot(days,temp_max,'r')
plt.plot(days,temp_min,'b')
```

```python
temp_min=list(np.random.randint(30,50,8))
temp_max=list(np.random.randint(30,50,8))
plt.scatter(days,temp_max,label='max')
plt.legend()
plt.axis()
```

```python
plt.colorbar(plt.scatter(days,temp_min,c=temp_min))
```
